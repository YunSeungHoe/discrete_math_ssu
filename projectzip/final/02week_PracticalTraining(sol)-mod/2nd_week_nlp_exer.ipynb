###  한글 형태소 분석기 사용해보기

```python
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Komoran

okt = Okt()
kkma = Kkma()
komran=Komoran()
text="열심히 코딩한 당신, 연휴에는 여행을 가봐요"

print('OKT 형태소 분석 :', okt.morphs(text))
print('kkma 형태소 분석 :', kkma.morphs(text))
print('komoran 형태소 분석 :', komran.morphs(text))
print('OKT 품사 태깅 :', okt.pos(text))
print('kkma 품사 태깅 :', kkma.pos(text))
print('komoran 품사 태깅 :', komran.pos(text))
print('OKT 명사 추출 :',okt.nouns(text)) 
print('kkma 명사 추출 :', kkma.nouns(text)) 
print('komoran 명사 추출 :', komran.nouns(text)) 
```

```python
import pandas as pd
import sentencepiece as spm
import urllib.request
import csv
```

네이버 영화 리뷰 데이터를 다운로드하여 판다 데이터프레임에 저장합니다.

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings.txt", filename="ratings.txt")
naver_df = pd.read_table('ratings.txt')
```

```python
print('리뷰 개수 :',len(naver_df)) # 리뷰 개수 출력
```

데이터 정제

```python
#네이버 영화 리뷰 데이터의 경우 Null 값이 존재하므로 이를 제거한 후에 수행합니다.

print(naver_df.isnull().values.any())

naver_df = naver_df.dropna(how = 'any') # Null 값이 존재하는 행 제거
print(naver_df.isnull().values.any()) # Null 값이 존재하는지 확인

print('리뷰 개수 :',len(naver_df)) # 리뷰 개수 출력
```
